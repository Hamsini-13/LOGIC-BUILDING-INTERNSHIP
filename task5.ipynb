```python
def analyze_post_engagement(likes_list):
    """
    Calculates total likes and determines if the post is viral.
    """
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example
analyze_post_engagement([200, 300, 250, 300])
```

```python
def check_medicine_stock(stock):
    """
    Checks if medicine stock is low or sufficient.
    """
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example
check_medicine_stock(6)
```

```python
def rainfall_checker(rainfall_list, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example
rainfall_checker([70, 80, 65, 75], 70)
```

```python
def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames using set.
    """
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)


# Example
detect_duplicate_accounts(["user1", "user2", "user3", "user1"])
```

```python
def appointment_eligibility(age):
    """
    Checks if patient is eligible based on age.
    """
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Example
appointment_eligibility(21)
```

```python
def premium_crop_filter(prices):
    """
    Filters crop prices above ₹2000.
    """
    premium = []
    
    # Filtering logic
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example
premium_crop_filter([1500, 2500, 3200, 1800])
```

```python
def application_health(errors):
    """
    Evaluates system health based on error count.
    """
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)


# Example
application_health(7)
```

```python
def transaction_checker(amount):
    """
    Validates transaction against ₹50,000 daily limit.
    """
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example
transaction_checker(60000)
```

```python
def attendance_checker(attendance_list):
    """
    Calculates attendance percentage and checks eligibility.
    """
    total_classes = len(attendance_list)
    attended = 0
    
    # 1 = Present, 0 = Absent
    for day in attendance_list:
        attended += day
    
    percentage = (attended / total_classes) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example
attendance_checker([1,1,1,1,0,1,1,1,0,1])
```

```python
def electricity_bill(units):
    """
    Calculates electricity bill based on slab rates
    and classifies usage level.
    """
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Classification:", usage)
    
    return bill, usage


# Example
electricity_bill(250)
```
