```python
# Problem 1: Smart Light Controller
# This class controls a smart light (ON/OFF)

class SmartLight:
    def __init__(self, light_name):
        # Initialize light name and default status as OFF
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        # Method to turn light ON
        self.status = "ON"

    def turn_off(self):
        # Method to turn light OFF
        self.status = "OFF"

    def display_status(self):
        # Display current status of light
        print(f"{self.light_name} is {self.status}")


# Sample Usage
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
# Problem 2: Employee ID Card System
# This class stores employee details and displays ID card

class Employee:
    def __init__(self, name, emp_id, department):
        # Initialize employee details
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        # Display employee ID card
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Usage
employee = Employee("Rahul", "EMP102", "IT")
employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record
# This class stores contact details

class Contact:
    def __init__(self, name, phone):
        # Initialize contact details
        self.name = name
        self.phone = phone

    def save_contact(self):
        # Display saved contact
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Usage
contact = Contact("Anita", 9876543210)
contact.save_contact()
```

```python
# Problem 4: Product Price Tag Generator
# This class generates a price tag for a product

class Product:
    def __init__(self, product_name, price):
        # Initialize product details
        self.product_name = product_name
        self.price = price

    def display_price_tag(self):
        # Print formatted price tag
        print("Product:", self.product_name)
        print(f"Price: ₹{self.price}")


# Sample Usage
product = Product("Headphones", 2499)
product.display_price_tag()
```

```python
# Problem 5: Movie Rating Display System
# This class stores movie details and rating

class Movie:
    def __init__(self, movie_name, rating):
        # Initialize movie details
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        # Display movie rating
        print("Movie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")


# Sample Usage
movie = Movie("Inception", 4.8)
movie.display_movie_details()
```
